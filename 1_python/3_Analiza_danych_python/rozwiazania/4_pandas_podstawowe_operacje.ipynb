## loc i iloc

Dobrą praktyką przy wyciąganiu danych z ramki jest używanie metod <i> loc </i> oraz <i> iloc </i>.

Metody te będą filtrować zarówno wiersze jak i kolumny. 
Do filtrowania możemy używać tych samych warunków, co w instrukacjach warunkowych.

* loc filtruje po wartościach wierszy i nazwach kolumn 
* iloc filtruje po numerach indeksów
```python
df.loc[wiersze]
df.loc[wiersze, kolumny]
df.iloc[wiersze]
df.iloc[wiersze, kolumny]
```
Alternatywnie: 


<b> Metody loc i iloc wymagają podania nawiasów kwadratowych</b>

<b>Dokumentacja </b>

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html

Kilka warunków w loc oddzielamy operatorami logicznymi:
- & (and)
- | (or)

Każdy warunek zapisujemy w nawiasie

````python
# Stany WA lub BC
df.loc[(df['State']=='WA') | (df['State']=='BC')]
````

````python
# Stan WA i miasto Wilkeson
df.loc[(df['State']=='WA') & (df['City'] =='Wilkeson')]
````

````python
# Wyciągnięcie jednego indeksu
df.iloc[4,:]
````

````python
# Wyciągnięcie po indeksie wiersza i kolumny
df.iloc[4,1]
````

````python
# Wyciągnięcie po indeksie wiersza i kolumny
df.iloc[4:,1:]
````

````python
# Liczba stanow
df.loc[:,"State"].nunique()
````

````python
# Liczba rekordów per stan
stany = df.loc[:,'State'].unique()
for stan in stany:
    records = len(df[df['State']==stan])
    print(f'Stan {stan} ma {records} rekordów w zbiorze.')
````

````python
# Pusta ramka danych
df_2 = pd.DataFrame()
df_2
````

````python
# Dodanie wartości 
df_2['State'] = ['WA','DC']
df_2
````

````python
# Zdefiniowane kolumny
df_2 = pd.DataFrame(columns = ['State','records'])
df_2
````

````python
# Ramka danych z liczbą rekordów per stan
stany = df.loc[:,'State'].unique()
i = 0 
for stan in stany:
    records = len(df[df['State']==stan])
    print(f'Stan {stan} ma {records} rekordów w zbiorze.')
    df_2.loc[i,:] = [stan, records]
    i+=1
````

````python
df_2 = pd.DataFrame(columns= ['State','records','count'])
````

````python
# Jeżeli w ramce jest więcej zmiennych, niż te, które modyfikujemy
stany = df.loc[:,'State'].unique()
i = 0 
for stan in stany:
    records = len(df[df['State']==stan])
    print(f'Stan {stan} ma {records} rekordów w zbiorze.')
    df_2.loc[i,'State'] = stan
    df_2.loc[i,'records'] = records
    i+=1
````
