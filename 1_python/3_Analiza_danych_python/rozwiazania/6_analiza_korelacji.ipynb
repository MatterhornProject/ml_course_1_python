# Korelacja
Korelacja obrazuje poziom zależności pomiędzy zmiennymi. 

Będziemy pracować na tym samym zbiorze danych.
https://www.kaggle.com/datasets/adarshde/electric-vehicle-population-dataset/data 

```python
import pandas as pd
```

```python
# puść ten kod, 
# jeżeli wywołujesz plik  w folderze rozwiąznaia, 
# a ramka danych znajduje się w folderze data
import os 
os.chdir('../')
```

```python
df = pd.read_csv("data/Electric_Vehicle_Population_Data.csv")
```

```python
# Przygotowanie danych numerycznych
df_numerical = df.loc[:,~(df.dtypes=='object')]
```

```python
## Korelacja Pearsona 
df_numerical.corr()
```

```python
## Korelacja Spearmana 
df_numerical.corr(method='spearman')
```

```python
# wybierzmy rekordy, gdzie BaseMSRP jest dodatni
base_msrp_positive = df[df['Base MSRP']>0].reset_index(drop = True)
```

```python
cols = df.columns[df.dtypes!='object']
```

```python
cols
```

```python
base_msrp_positive.shape
```

```python
## Korelacja Pearsona 
base_msrp_positive[cols].corr()
```

```python
## Korelacja Spearmana 
base_msrp_positive[cols].corr(method = 'spearman')
```

```python
electric_range_positive = df[df['Electric Range']>0].reset_index(drop = True)
```

```python
electric_range_positive[cols].corr(method = 'spearman')
```

```python
electric_range_positive.groupby('Model Year').count()
```

```python
electric_range_positive[(electric_range_positive['Model Year']>=2011) & (electric_range_positive['Model Year']<2025)][cols].corr(method = 'spearman')
```

```python
electric_range_positive.head()
```

```python
electric_range_positive[(electric_range_positive['Model Year']>=2011) & (electric_range_positive['Model Year']<2025)].groupby('Electric Vehicle Type')[cols].corr(method = 'spearman')
```
