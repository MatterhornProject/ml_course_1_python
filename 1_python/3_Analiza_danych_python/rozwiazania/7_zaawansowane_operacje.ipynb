## Metoda apply

Metoda do aplikowania zadanej funkcji dla poszczególnych wierszy lub kolumn

Dokumentacja: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html

```python
# Metoda apply na kolumnach
df.apply(np.sum)
```

```python
# Metoda apply na wierszach
df.apply(np.sum, axis = 1)
```

```python
df.describe()
```

```python
# Przeliczenie nieobecności na godziny
df['hourly_absence'] = df['Absences'] * 8
```

```python
# Suma godzin nieobecności i pracy własnej
df['sum_absence_and_study'] = df[['StudyTimeWeekly','hourly_absence']].apply(np.sum, axis = 1 )
```

```python
df['sum_absence_and_study']
```

```python
# Apply w groupby
df.groupby('GradeClass').apply(np.sum, axis=0, include_groups=False)
```

## Funkcja lambda

Funkcja lambda służy do tworzenia krótkich funkcji, które mają tylko jedną operację.

```python
# Lambda z jednym argumentem
x = lambda y: y **3
```

```python
# Wywołanie
x(10)
```

```python
# Lambda z wieloma argumentami

func_lambda= lambda x,y: y if x<0 else x 
```

```python
func_lambda(-2,0)
```

### Funkcja lambda w apply

```python
# Dodanie zmiennej z informacją, czy są jakieś nieobecności
df['are_absences'] = df['Absences'].apply(lambda x: 'absences' if x>0 else 'no absences')
```

```python
df['are_absences']
```

```python
df.are_absences.value_counts()
```

```python
# Suma nieobecności i godzin samodzielnej pracy
df['hourly_absence_and_working'] = df.loc[:,['Absences','StudyTimeWeekly']].apply(lambda x: x.iloc[0]*8 + x.iloc[1], axis = 1)
```

```python
del df['hourly_absence_2']
```

```python
df.head()
```

```python
# Własna funkcja 
# Napiszemy funkcję do kategoryzacji liczby nieobecności
def absence_kat(absence):
    if absence <=0:
        return 'no absences'
    elif absence <=15:
        return 'Acceptable amount of absences'
    else:
        return 'High number of absences'
```

```python
#Stworzenie zmiennej z kategorią nieobecności
df['absences_kat'] = df['Absences'].apply(lambda x: absence_kat(x))
```

```python
df.head()
```

```python
# Przykład użycia lambdy w group by
# Załóżmy, że średnie są zawyżone o 1. Wyliczymy poprawne średnie po kategorii nieobecności
df.groupby('absences_kat')['GPA'].apply(lambda x: np.mean(x-1))
```

## Metoda rolling
Metoda do wyliczania metryk w oparciu o bieżący rekord i n wcześniejszych.

Dokumentacja: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rolling.html

```python
# Wykorzystamy stworzoną wcześniej, zagregowaną ramkę danych 
df_grouped_electrics_cars = pd.read_csv('data/df_by_year.csv')
```

```python
df_grouped_electrics_cars = df_grouped_electrics_cars.sort_values(by = 'Model Year')
```

```python
# Średnia z dwóch ostatnich lat
df_grouped_electrics_cars['Electric Range'].rolling(window=2).mean()
```

```python
# Średnia z dwóch ostatnich lat z wypełnioną pierwszą wartością
df_grouped_electrics_cars['Electric Range'].rolling(window=2, min_periods=1).mean()
```

```python
# Wykorzystamy stworzoną wcześniej, zagregowaną ramkę danych po latach i producencie 
df_grouped_electrics_cars_2 = pd.read_csv('data/df_by_year_make.csv')
df_grouped_electrics_cars_2 = df_grouped_electrics_cars_2.sort_values(by = ['Make','Model Year'])
```

```python
# Rolling window w grupach po producencie
df_grouped_electrics_cars_2[['Electric Range','Make']].groupby('Make').rolling(window=2, min_periods=1).mean()
```
